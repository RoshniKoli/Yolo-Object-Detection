#### Converting txt file from dos to unix format

```python
train_txt = glob.glob(r'C:\Users\roshn\Documents\DataAnalytics\Springboard\CV_Fashion_dataset\train\train\YOLO_train\*.txt')
```

```python
WINDOWS_LINE_ENDING = b'\r\n'
UNIX_LINE_ENDING = b'\n'

for file in train_txt:
    with open(file,'rb') as open_file:
        content = open_file.read()
    content = content.replace(WINDOWS_LINE_ENDING, UNIX_LINE_ENDING)
    with open(file,'wb') as open_file:
        open_file.write(content)
```
